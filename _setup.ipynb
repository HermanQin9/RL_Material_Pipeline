```python
import os

# 要创建的目录列表
dirs = [
    'methods',
    'data/raw',
    'data/processed',
    'models',
    'logs',
    'dash_app/data'
]

# 要创建的空文件列表
files = [
    'pipeline.py',
    'nodes.py',
    'methods/data_methods.py',
    'methods/model_methods.py',
    'env.py',
    'train_ppo.py'
]

# 创建目录
for d in dirs:
    os.makedirs(d, exist_ok=True)

# 创建空文件
for f in files:
    open(f, 'a').close()

print("Project structure created.")
```

```python
# 简单地打印当前目录下的两级树状结构
def print_tree(path, prefix=''):
    files = sorted(os.listdir(path))
    pointers = ['├── '] * (len(files) - 1) + ['└── ']
    for pointer, name in zip(pointers, files):
        print(prefix + pointer + name)
        full_path = os.path.join(path, name)
        if os.path.isdir(full_path):
            extension = '│   ' if pointer == '├── ' else '    '
            print_tree(full_path, prefix + extension)

import os
print_tree('.')
```
